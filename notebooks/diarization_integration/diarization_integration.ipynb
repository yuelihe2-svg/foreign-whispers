````python
import wave
from pathlib import Path
import sys

PROJECT_ROOT = Path.cwd().parent.parent

# 我们刚才提取的完整视频音频路径
source_wav = PROJECT_ROOT / "pipeline_data" / "api" / "diarizations" / "Strait of Hormuz disruption threatens to shake global economy.wav"

# 目标音色库路径
spk_dir = PROJECT_ROOT / "pipeline_data" / "speakers" / "es"
spk_dir.mkdir(parents=True, exist_ok=True)

if source_wav.exists():
    with wave.open(str(source_wav), 'rb') as src:
        params = src.getparams()
        framerate = params.framerate

        # 从原音频中截取 3 段真实的原始声音（跳过开头无声部分，每段切 3 秒）
        start_times = [10, 30, 135] # 分别从第10秒、第30秒、第135秒开始切
        
        for i, start_s in enumerate(start_times):
            src.setpos(framerate * start_s) 
            frames = src.readframes(framerate * 3) # 读取 3 秒钟的数据
            
            out_path = spk_dir / f"voice_{i}.wav"
            with wave.open(str(out_path), 'wb') as out:
                out.setparams(params)
                out.writeframes(frames)
            print(f"✅ 成功截取真实原声样本: {out_path.name} (来自原视频的第 {start_s} 秒)")
else:
    print("❌ 未找到原音频文件，请确认路径。")
````
